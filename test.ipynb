```python
import json
import pprint
import pandas as pd
```

```python
with open("./Raw_data/2018/201800000") as file: 
    data = json.load(file)  # Parse the JSON content into a Python dictionary
    # pprint.pprint(data.get('abstracts-retrieval-response').get('item'))  # Access the key
    pprint.pprint(data.get('abstracts-retrieval-response').get('item').get('bibrecord').get('tail'))  # Access the key
```

```python
def print_json_keys(data, parent_key=''):
    """
    Recursively prints all keys in a nested JSON object.
    If the value of a key is a dictionary, it goes deeper into the dictionary.
    """
    if isinstance(data, dict):  # If the data is a dictionary
        for key, value in data.items():
            new_key = f"{parent_key}.{key}" if parent_key else key
            print(new_key)  # Print the key
            print_json_keys(value, new_key)  # Recursively go deeper
    elif isinstance(data, list):  # If the data is a list
        for i, item in enumerate(data):
            new_key = f"{parent_key}[{i}]"
            print_json_keys(item, new_key)  # Recursively go deeper into each list item
```

```python
print_json_keys(data)  # Call the function with the data
```
